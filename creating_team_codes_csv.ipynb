```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv

url = "https://fbref.com/en/squads/"

resp = requests.get(url).text.replace('<!--', '').replace('--!>', '')

soup = BeautifulSoup(resp)
club_count = soup.find_all('td', {'data-stat':'club_count'})

countries, urls, counts = [], [], []

for club in club_count:
    countries.append(club.a.get('href').split('/')[4])
    urls.append(f"https://fbref.com{club.a.get('href')}")
    counts.append(int(club.get_text()))
    
counts_df = pd.DataFrame(list(zip(countries, counts, urls)), columns = ['Country', 'Count', 'Url'])
counts_df = counts_df.sort_values(by = 'Count', ascending = False)

relevant_countries = counts_df.head(10)

team_codes = {}

for i, row in relevant_countries.iterrows():
    country_url = row['Url']
    resp = requests.get(country_url).text.replace('<!--', '').replace('--!>', '')
    soup = BeautifulSoup(resp)
    teams = soup.find_all('th', {'data-stat':'team'})
    
    squad_ids, squads = [], []
    
    for team in teams:
        try:
            squad_ids.append(team.a.get('href').split('/')[3])
            squads.append(team.get_text())
        except: pass
        
    new_team_codes = dict(zip(squad_ids, squads))
    
    team_codes.update(new_team_codes)
    
with open('team_codes.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Team_ID', 'Team'])
    for key, value in team_codes.items():
        writer.writerow([key, value])
```
